## Stop Words in NLTK

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
example_text ="Hello, I am a human. I am from Nepal, a beautiful country in laps of Himalayas."
another_sample_text= "The quick brown fox jumps over the lazy dog. It is a very common sentence used in the English language. The purpose of this sentence is to showcase all the letters of the alphabet."
```

```python
stop_words = set(stopwords.words("english"))
stop_words
```

```python
sen1_words = word_tokenize(example_text)
sen2_words = word_tokenize(another_sample_text)
fileterd_sen1 =[]
fileterd_sen2 =[]

for sig_word in sen1_words:
    if sig_word not in stop_words:
        fileterd_sen1.append(sig_word)


for sig_word in sen2_words:
    if sig_word not in stop_words:
        fileterd_sen2.append(sig_word)

print(fileterd_sen1)
print(fileterd_sen2)
```

```python
fileterd_sen3 = [ w for w in sen1_words if w not in stop_words]
fileterd_sen3
```
